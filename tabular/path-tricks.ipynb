```python
def _d_repr(d): return color.BOLD+color.BLUE+str(d)+color.END+'\n'
def _f_repr(f): return str(f)+'\n'
```

```python
def walk2dict(root):
    d = {}
    root = Path(root)
    for path, dirs, files in walk(root):
        path = Path(path).relative_to(root)
        parent = reduce(dict.get, path.parts[:-1], d)
        parent[path.name] = dict.fromkeys(files)
    if '' in d: d = {**d, **d.pop('')} # TODO: fix earlier?
    return d
```

```python
def _ptree(pdict, level=0, cons_lasts=0):
    n = len(pdict)
    for i, (k, vs) in enumerate(pdict.items()):
        last = i == n-1
        s = '└' if last else '├'
        pre = ''
        if level <= 0: pre += s+'─'
        for _ in range(level - cons_lasts): pre += '│   '
        for _ in range(cons_lasts): pre += '    '
        if level > 0: pre += s+'─'
        pre += '─ '
        if k == '': k = '.'
        if vs is None:
            p.text(pre+_f_repr(k))
        else:
            p.text(pre+_d_repr(k))
            if last: cons_lasts +=1
            _ptree(vs, level+1, cons_lasts)

def ptree(path):
    p.text(_d_repr(path))
    pdict = walk2dict(path)
    _ptree(pdict)
```
